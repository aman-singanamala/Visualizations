### Line Plot

```python
from turtle import color
import numpy as np
import matplotlib.pyplot as plt
X= np.arange(1,11)
y=2*X
z=3*X
plt.plot(X,y,color='g',linestyle=':',linewidth=2)
plt.plot(X,z,color='r',linestyle='--',linewidth=2)
plt.title("Line plot")
plt.xlabel("X-label")
plt.ylabel("y-label")
plt.grid(True)
plt.show()
```
